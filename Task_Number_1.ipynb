```python
"""
Импортируем все необходимые библиотеки
"""
import math
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import mlab
from scipy.integrate import odeint
```

```python
"""
Решим дифференциальное уравнение второго порядка с помощью функции odeint, определенной в библиотеке scipy
Для начала определим нашу систему в функции diff_equaion(U, x), где U - это вектор: y = U[0], z = U[1]
Затем передадим начальные условия и запустим odeint, а также построим наши графики в функции plot_dif_equation()
"""
def diff_equation(U, x):
    return [U[1], -100 * U[0] - 101 * U[1]]


def plot_dif_equation(x_axis, init_conditions):
    solver = odeint(diff_equation, init_conditions, x_axis)
    y_axis = solver[:,0]
    
    
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Second order diff equation by Shamil")
    plt.plot(x_axis,y_axis);
```

```python
"""
Определим начальные условия для построения последующих графиков
"""
init_conditions = [2, -2]
```

```python
x_axis = np.linspace(0, 100, 2)
plot_dif_equation(x_axis, init_conditions)
```
