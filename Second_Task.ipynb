## Solve a differential equation:
# Выполнил: Вагабов Шамиль

```python
import numpy as np
from scipy.integrate import solve_bvp
import pylab as plt
```

```python
def equation(x, y):
    return np.vstack((y[1], (10 + x ** 2) * y[0] + x * np.exp(-x)))
```

```python
def boundary_conditions(ya, yb):
    return np.array([ya[0] - 3, yb[0] - np.exp(2)])
```

```python
def solve_equation(step):
    x = np.arange(0, 10, step)
    y = np.zeros((2, x.size))
    return solve_bvp(equation, boundary_conditions, x, y)
```

```python
def plot_equation(step):
    plt.figure(figsize=(5, 5))
    x_ax = np.arange(0, 10, step)
    plt.plot(x_ax, solve_equation(step).sol(x_ax)[0], label="y")
    plt.xlabel('x')
    plt.ylabel('y')
```

```python
plot_equation(1e-1)
```

```python
plot_equation(1e-2)
```

```python
plot_equation(1e-3)
```

```python
plot_equation(1e-4)
```
