```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

```python
def euler_method(y, z, h, N):
    for i in range(0, N-1):
        y[i+1] = (y[i] + z[i]*h)
        z[i+1] = (y[i]*(-100*h) + z[i]*(1-101*h))
```

```python
def plot(i, h, dig):
    counter = int(10/h + 1)
    x = np.linspace(0, 10, counter)
    
    counter = int(10/h + 1)
    y = np.zeros(counter)
    z = np.zeros(counter)
    y[0] = 2
    z[0] = -2
    
    plt.figure(figsize=(18, 12))
    graphics = plt.subplot(4, 2, i + 1)
    euler_method(y, z, h, counter)
    plt.plot(x, y)
    graphics.set_title("Шаг равен {0}".format(h))
    
h_min = 2e-2
h_max = 3e-2
dig = 4
for i in range(4):
    plot(i, h_min + i*(h_max-h_min)/(dig-1), dig+1)
```

# При шаге большем 0.02 неустойчива

```python
w, v = LA.eig(np.array([[0, 1], [-100, -101]]))
for i in range(len(w)):
    w[i] = abs(w[i])
result = max(w) / min(w)
print (result)
```
